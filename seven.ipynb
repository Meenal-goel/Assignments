```python
#5.Multivariate Analysis 
#Plot the multivariate graphs and find the relationship between the variables. 

#plot 1:realtions between all features of the dataset
import pandas as pd
import seaborn as sns
dff = pd.read_csv("C:/Users/hp/Desktop/Machine_Learning/Assignments/kc_house_data.csv")
x = dff.columns
cor_mat = dff[x].corr()
sns.heatmap(cor_mat)
```

```python
#INFERENCE : The id feature is perfectly co related with all other features of this dataset\
#with highest corelation value i.e 1
'''From the corrmat we can see that the variables bathrooms, bedrooms, floors, sqft_living,\
grade are highly correlated. The variable sqft_above, sqft_basement are also correlated but we will not include \
it in the analysis because all the information is encapsulated in sqft_living. The sqft_living15 is also correlated\
but we will not include it in the analysis as it is gives us the same information as sqft_living. \
All the other variables are weakly correlated.'''
```
