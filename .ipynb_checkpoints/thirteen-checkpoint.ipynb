```python
#1.) Import the dataset from sklearn. 
from sklearn.datasets import load_digits
digits = load_digits()


#Q.2- Split the dataset into training and testing sets using train_test_split() 
from sklearn.model_selection import train_test_split

x =  digits.data[:-10]
y = digits.target[:-10]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3,random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
regressor = LogisticRegression()
regressor.fit(x_train,y_train)
print("y_intercept:",regressor.intercept_)
print("slope:",regressor.coef_)

print("*"*110)

#prediction
y_pred = regressor.predict(x_test)
dataset = pd.DataFrame({"actual":y_test,"predicted":y_pred})
print(dataset)

```

```python
from sklearn import metrics
#classification accuracy
metrics.accuracy_score(y_test, y_pred)
```

```python
#logarithmic loss
pred = regressor.predict_proba(x_test)
 metrics.log_loss(y_test, pred)
```

```python
#r2 score
x =  digits.data[:-10]
y = digits.target[:-10]
metrics.r2_score(y_test, y_pred)
```

```python
#MAE AND MSE
print("MAE",metrics.mean_absolute_error(y_test,y_pred))
print("MSE",metrics.mean_squared_error(y_test,y_pred))
```

```python
#confusion matrix and classification report
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
