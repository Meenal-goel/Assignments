```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
dff = pd.read_csv("C:/Users/hp/Desktop/Machine_Learning/Assignments/weather2.csv")
#UNIVARIATE ANALYSIS ON NUMERICAL DATA
sns.distplot(dff["MinTemp"])#by default hist = True ,kde = True
```

```python
sns.distplot(dff["MaxTemp"], hist = True, kde = False)# to display only histogram
```

```python
#using rug attribute to display density of datapoints
sns.distplot(dff["Evaporation"],hist = True,kde =False, rug = True)
```

```python
sns.distplot(dff["WindSpeed9am"],hist = False,kde =True, rug = True)
```

```python
#to shade the area covered by the kde
sns.kdeplot(dff["WindSpeed9am"],shade = True)
```

```python
#to analyse two different data columns in same plot(to show comparison)
sns.kdeplot(dff["Humidity9am"],label="Humidity9am")
sns.kdeplot(dff["Humidity3pm"],label="Humidity3pm")
```

```python
#binning
sns.kdeplot(dff["Temp9am"],bw = 0.2, label = "Temp9am" )
sns.kdeplot(dff["Temp3pm"], bw = 0.5, label = "Temp3pm" )


```

```python
#to make scattered plot
dff.plot(kind = "scatter", x = "MinTemp", y = "MaxTemp")
```

```python
sns.jointplot(x = "Pressure9am", y = "Pressure3pm", data = dff)
```

```python
#to clearly differ densities of datapoints
sns.jointplot(x = "Pressure9am", y = "Pressure3pm", kind ="hex", data = dff)
```

```python
#to clearly differ densities of datapoints using color attribute
sns.jointplot(x = "Pressure9am", y = "Pressure3pm", kind ="hex", data = dff, color ="k")
```

```python
with sns.axes_style("white"):
    sns.jointplot( x ="MinTemp", y = "MaxTemp", data = dff,kind = "kde")
```

```python
#2D Density plot
sns.jointplot(x ="Cloud9am", y ="Cloud3pm", data = dff, kind = "kde")
```

```python
#pair plot
iris = sns.load_dataset("iris")
sns.pairplot(iris)
```

```python
#CATEGORICAL DATA
sns.stripplot(x="WindGustDir", y="Rainfall", data = dff)
```

```python
#use jitter to clearly identify density
sns.stripplot(x="WindGustDir", y="Rainfall", data = dff, jitter = True)
```

```python
#Swarm plot
sns.swarmplot(x = "WindGustSpeed", y="WindDir9am", data = dff)
```

```python
#box plot
sns.boxplot(x = "WindGustSpeed", y="WindDir9am", data = dff)
```

```python
#Bar plot
sns.barplot(x="Date", y="WindSpeed9am", data = dff)
```

```python
#Count plot
sns.countplot(x ="WindSpeed9am",data =dff)
```
