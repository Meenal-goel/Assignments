```python
#1) Import the required libraries (sklearn, pandas, numpy). Import datasets. Concatenate the datasets.\
#Display first 5 rows.

#import required libraries
import pandas as pd
import numpy as np


#import the two datasets
d1 = pd.read_csv("C:/Users/hp/Desktop/Machine_Learning/Assignments/train.csv")
d2 = pd.read_csv("C:/Users/hp/Desktop/Machine_Learning/Assignments/test.csv")

#concatenate the two datasets
dff_ = d1.append(d2, sort = False)

#display first 5 rows of the dataset
print("FIRST 5 ROWS OF THE DATASET")
print(dff_.head())
print("*"*110)

#2) Remove NaN values from the column ‘Survived’. (Hint: use .dropna())
print("NOT NULL VALUES :",dff_["Survived"].notnull().sum())
print("NULL VALUES :",dff_["Survived"].isnull().sum())
#DROPPING NULL VALUES
dff_["Survived"].fillna(dff_["Survived"].mode()[0], inplace = True)
print("NUMBER OF NULL VALUES AFTER USING DROPNA:",dff_["Survived"].isnull().sum())
```

```python
#3.Create a new column 'Has_Cabin' that tells you whether passengers had a cabin or not. 
#(Hint: use .isnull() method in “Cabin” column) 

dff_["Has_Cabin"] = np.where(dff_["Cabin"].isnull() == True ,"no", "yes")
dff_.head()
```

```python
#4) Create a new column “Title”, which will contain the titles of the passengers such as Mr., Mrs., etc.\
#There should be five titles: Mr, Mrs, Master, Miss, and Special. 
#(Hint: Use regular expression to replace the titles)
import re
dff_['Title'] = dff_.Name.apply(lambda x: re.search(' ([A-Z][a-z]+)\.', x).group(1))
dff_['Title'] = dff_['Title'].replace({'Mlle':'Miss', 'Mme':'Mrs', 'Ms':'Miss'})
dff_['Title'] = dff_['Title'].replace(['Don', 'Dona', 'Rev', 'Dr',
                                            'Major', 'Lady', 'Sir', 'Col', 'Capt', 'Countess', 'Jonkheer'],'Special')
dff_
```

```python
#5. Impute the missing values using .fillna() method. 
dff_.isnull().sum()
```

```python
#5. Impute the missing values using .fillna() method. 
for column in ["Age","Fare"]:
    dff_[column].fillna(dff_[column].median(), inplace = True)
for column in ["Cabin","Embarked"]:
    dff_[column].fillna(dff_[column].mode()[0], inplace = True)
dff_.isnull().sum()
```

```python
#6) Create two new columns “Categorical_age” and “Categorical_fare”, where you will put people \
#that are within a certain range of age or fare in the same bin.
dff_['Categorical_Age'] = pd.qcut(dff_.Age, q=4, labels=False )
dff_['Categorical_Fare']= pd.qcut(dff_.Fare, q=4, labels=False)
dff_.head()
```

```python
#7) Drop columns Cabin, Name, Passengerid, Ticket, Age, Fare, SibSp, Parch.
dff_ = dff_.drop(["Cabin", "Name", "PassengerId", "Ticket", "Age", "Fare", "SibSp", "Parch"], axis=1)
dff_.head()
```

```python
#8) Transform all variables into numeric ones using get_dummies().
dff_dummy = pd.get_dummies(dff_, drop_first=True)
dff_dummy.head()
```

```python
#9) Now we need training set and testing set. Use train_test_split() to split your dataset. 
#X(Features) will be the dataset values without the column “Survived” and y(Targets) will be your “Survived” \
#column values.
from sklearn.model_selection import train_test_split
x = dff_[["Pclass"]]
y = dff_["Survived"]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3,random_state = 0)


#10) Use logistic regression by passing your X_train and y_train.
from sklearn.linear_model import LogisticRegression
regressor = LogisticRegression()
regressor.fit(x_train,y_train)
print("y_intercept:",regressor.intercept_)
print("slope:",regressor.coef_)

```

```python
#prediction
y_pred = regressor.predict(x_test)
dataset = pd.DataFrame({"actual":y_test,"predicted":y_pred})
print(dataset)
```

```python
#11) Print the accuracy score.
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred)
```
